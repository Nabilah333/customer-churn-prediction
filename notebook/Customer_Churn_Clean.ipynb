```python
## Feature Note

# The CLTV feature was retained because the dataset documentation states that it is a predicted Customer Lifetime Value derived from existing customer data. Since it represents a business metric that companies commonly compute before customer churn occurs, it is treated as an available feature rather than data leakage.
```

```python
df["Total Charges"] = pd.to_numeric(
    df["Total Charges"],
    errors="coerce"
)
```

```python
df["Total Charges"] = df["Total Charges"].fillna(0)
```

```python
df["Total Charges"].isnull().sum()
```

```python
columns_to_drop = [
    "CustomerID",
    "Count",
    "Country",
    "State",
    "Churn Label",
    "Churn Reason",
    "Churn Score"
]

df = df.drop(columns=columns_to_drop)
```

```python
df.info()
```

```python
df.to_excel(
    "Telco Customer Churn Cleaned.xlsx",
    index=False
)
```
